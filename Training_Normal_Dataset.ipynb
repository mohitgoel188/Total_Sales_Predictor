# Imports

```python
import matplotlib.pyplot as plt
from beakerx import *
import pandas as pd
import numpy as np
%matplotlib inline
```

# Data Analysis

```python
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
```

```python
train.info()
```

```python
train.describe()
```

```python
train.head(5)
```

```python
train.drop(['store_location','city','state','time_zone','location_employee_code',
            'credit_score','credit_score_range','outlet_no','latitude','longitude'],axis=1,inplace=True)
train.head()
```

```python
test.drop(['store_location','city','state','time_zone','location_employee_code',
            'credit_score','credit_score_range','latitude','longitude'],axis=1,inplace=True)
test.head()
```

# Train Test Split

```python
train.head()
```

```python
X=train.drop('total_sales',axis=1)
Y=train['total_sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, Y, test_size=0.3, random_state=101)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
XS_train=scaler.transform(X_train)
```

```python
XS_val=scaler.transform(X_val)
```

# Training


```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import ElasticNet
```

```python
param_grid={'alpha':[0.03,0.1,0.3,1,3,10,30,100,300],'l1_ratio':[0,0.01,0.02,0.03,0.04,0.045],'max_iter':[5,10,100]}
```

```python
elastic=GridSearchCV(ElasticNet(),param_grid,verbose=2)
```

```python
elastic.fit(X_train,y_train)
```

```python
elastic.best_estimator_
```

# Validating

```python
from sklearn import metrics
```

```python
y_pred=elastic.predict(X_val)
```

```python
print(np.sqrt(metrics.mean_squared_error(y_val,y_pred)))
```

```python
y_pred=elastic.predict(XS_val)
```

```python
print(np.sqrt(metrics.mean_squared_error(y_val,y_pred)))
```

# Testing

```python
X_test=test.drop('outlet_no',axis=1)
```

```python
y_pred=elastic.predict(X_test)
```

```python
y_pred=y_pred.astype(int)
```

```python
ans=pd.DataFrame()
ans['outlet_no']=test['outlet_no']
ans["total_sales_actual"]=y_pred
```

```python
ans.to_csv('ans_n.csv',index=False)
```
