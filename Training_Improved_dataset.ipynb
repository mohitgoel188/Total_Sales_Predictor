# Imports

```python
import matplotlib.pyplot as plt
from beakerx import *
import pandas as pd
import numpy as np
%matplotlib inline
```

# Load Cleaned Dataset


```python
train=pd.read_csv('trainX.csv')
test=pd.read_csv('testX.csv')
```

# Train Test Split

```python
train.head()
```

```python
X=train.drop(['total_sales'],axis=1)
Y=train['total_sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, Y, test_size=0.3, random_state=101)
```

# Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
XS_train=pd.DataFrame(data=scaler.transform(X_train),index=X_train.index,columns=X_train.columns)
```

```python
XS_val=XS_train=pd.DataFrame(data=scaler.transform(X_val),index=X_val.index,columns=X_val.columns)
```

# Training


```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import ElasticNet
```

```python
param_grid={'alpha':[0.1,0.3,1,3,10,30,100,300,500],'l1_ratio':[0,0.01,0.04,0.1,0.25,0.5,0.75,1],'max_iter':[5,10,100,1000]}
```

```python
grid=GridSearchCV(ElasticNet(),param_grid,verbose=2)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_estimator_
```

# Validating

```python
from sklearn import metrics
```

```python
y_pred=grid.predict(X_val)
```

```python
print(np.sqrt(metrics.mean_squared_error(y_val,y_pred)))
```

```python
y_pred=grid.predict(XS_val)
```

```python
print(np.sqrt(metrics.mean_squared_error(y_val,y_pred)))
```

# Testing

```python
#X_test=scaler.transform(test.drop('outlet_no',axis=1))
X_test=test.drop(['outlet_no'],axis=1)
```

```python
y_pred=grid.predict(X_test)
```

```python
y_pred=y_pred.astype(int)
```

```python
ans=pd.DataFrame()
ans['outlet_no']=test['outlet_no']
ans["total_sales_actual"]=y_pred
```

```python
ans.to_csv('ans_dc.csv',index=False)
```
